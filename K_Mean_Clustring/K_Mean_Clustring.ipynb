## K Means Clustring Alogrithim

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## STandarization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train_Scaled=scaler.fit_transform(X_train)
X_test_Scaled=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow Method TO Select K Value
wscc=[]
for k in range(1,11):
    kmean=KMeans(n_clusters=k,init="k-means++")
    kmean.fit(X_train_Scaled)
    wscc.append(kmean.inertia_)
```

```python
wscc
```

```python
## Plot elbow Curve
plt.plot(range(1,11),wscc)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmean=KMeans(n_clusters=3,init="k-means++")
```

```python
kmean.fit_predict(X_train_Scaled)
```

```python
y_pred=kmean.predict(X_test_Scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
## silhoute score
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeff=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=4,init="k-means++")
    kmeans.fit(X_train_Scaled)
    score=silhouette_score(X_train_Scaled,kmeans.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_coeff
```

```python
plt.plot(range(2,11),silhouette_coeff)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhoute Coefficient")
```
