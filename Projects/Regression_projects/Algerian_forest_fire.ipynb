## Algerian Forest Fires Dataset
Data Set Information:

The Dataset inculdes 244 instance that group of two region of Algeria,namely the Bejaia region located in the northeast of Algreian and the sidi Bel-abbes Region located in the northwest of Algeria.

122 instance for each region.

The period from June 2012 to september 2012. The Dataset inculdes 11 and 1 output  atrribute(class).The 244 instance have been classified into fire(138 classes) and not fire(106 classes) classes. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## Missing Value
dataset[dataset.isnull().any(axis=1)]
```

The Dataset is convereted into two sets based on Region from 122th index,we can make a new columns based on the Region

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes region Dataset"

Add new columns with region 

```python
dataset.loc[:122,"region"]=0
dataset.loc[122:,"region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['region']]=df[['region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Changes the required columns as Integer Data Type

```python
df[['day','month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day','month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other columns to float datatype
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
       df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_dataset_Cleaned.csv',index=False)
```
