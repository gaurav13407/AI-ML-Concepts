## Data Cleaning

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[-df['Reviews'].str.isnumeric()]
```

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[-df_copy['Reviews'].str.isnumeric()]
```

```python
## Convert review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000',regex=False)
df_copy['Size']=df_copy['Size'].str.replace('k','',regex=False)
df_copy['Size']=df_copy['Size'].replace(['Varies with device',''],np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df.head()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for items in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(items,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Installs'].astype(float)
```

```python
df_copy.info()
```

```python
### Handling the last update feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy.to_csv('data/googlecleaned.csv')
```

## Explore data More

```python
numeric_feature=[feature for feature in df.columns if df_copy[feature].dtype!='O']
categorical_feature=[feature for feature in df.columns if df_copy[feature].dtype=='O']
## Print colums
print('We have {} numeric features:{}'.format(len(numeric_feature),numeric_feature))
print('We have {} categorical features:{}'.format(len(categorical_feature),categorical_feature))
```

```python
## Porportion of count data on categorical value
for col in categorical_feature:
    print(df[col].value_counts(normalize=True)*100)
    print('--------------------------------------')
```

```python
## Porpotion of count data on categorical data on categorical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis Of numerical Feature',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numeric_feature)):
    plt.subplot(5,3, i+1)
    sns.kdeplot(x=df_copy[numeric_feature[i]],shade=True,color='r')
    plt.xlabel(numeric_feature[i])
    plt.tight_layout()
```

## Observation 
Rating and Year if left skewd while review,size,Installs and Price are right skewed

```python
## Categorical columns
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis Of Categorical Feature',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
categorical=['Type','Content Rating']

for i in range(0,len(categorical)):
    plt.subplot(5,3, i+1)
    sns.countplot(x=df[categorical[i]],palette="Set2")
    plt.xlabel(categorical[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
