## Decision Tree Classifier Practicle Implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
##Independent fearure
X=pd.DataFrame(iris['data'],columns=['speal lenght','speal widht','petal lenght','petal width'])
```

```python
## Dependent Features
y=iris['target']
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
## Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## Visulaize The Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_pred,y_test)
print(cm)
print(classification_report(y_pred,y_test))
```

## Decision Tree Prepruning and Hperparameter Tunning

```python
params={
    'criterion':["gini", "entropy", "log_loss"],
    'splitter':["best", "random"],
    'max_depth':[1,2,3,4,5],
    'max_features':[int, float or "sqrt", "log2"]
    
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(treemodel,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
cm=confusion_matrix(y_pred,y_test)
print(cm)
print(classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
print(score)
```
