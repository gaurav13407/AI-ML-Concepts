## Data Visualization With Seaborn

Seaborn is a python visualization library based on matplotlib that provide a high_level interface for drawing attractive and informative statistical graphics.Seaborn helps in creating complex visualtization with a few lines of code .

```python
import seaborn as sns
```

```python
## Basic plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
## Create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plot of total bill")
plt.show()
```

```python
## Line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title(" Size plot wrt tip")
plt.show()
```

```python
### Categorial  plots
## bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("plot of bar")
plt.show()
```

```python
## Box PLot
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
## Violin plot

sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
## Histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## PAirplot 

sns.pairplot(tips)
```

```python
## Heat map

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
