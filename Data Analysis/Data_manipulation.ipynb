```python
## Handling missing value
df.isnull().any()
```

```python
df_filled=df.fillna(0)
```

```python
## Filling missing value with a mean of the column
df['website_fillna']=df['website'].fillna(df['website'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming columns
df=df.rename(columns={'Name':'ID-23'})
df.head()
```

```python
## Change data type
df['value_ner']=df['Founded'].astype(int)
df.head()
```

```python
##
df['New_value']=df['Founded'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating and Grouping
df.head()
```

```python
grouped_mean=df.groupby('Number of employees')['Founded'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Founded','Industry'])['Value'].sum()
print(grouped_sum)
```

```python
## Aggregate multiple function
gropued_agg=df.groupby('Founded')['Industry'].agg(['mean','sum','count'])
gropued_agg
```

```python
## Merging and joining Dataframes
## Create sample Dataframe
df1=pd.DataFrame({'keys':['A','B','C'],'Vlaue1':[1,2,3]})
df2=pd.DataFrame({'keys':['A','B','D'],'Vlaue2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge Dataframe on the 'Key Columns'
pd.merge(df1,df2,on="keys",how="inner")
```

```python
pd.merge(df1,df2,on="keys",how="outer")
```

```python
pd.merge(df1,df2,on="keys",how="left")
```

```python
pd.merge(df1,df2,on="keys",how="right")
```
