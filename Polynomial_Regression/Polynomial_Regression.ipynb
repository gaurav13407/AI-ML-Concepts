## Polynomial Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=6*np.random.rand(100,1)-3
Y=0.5*X**2+1.5*X+2+np.random.randn(100,1)
##Quadrtic equation used y=0.5x^2+1.5x+2+outliers
plt.scatter(X,Y,color='g')
plt.xlabel('X dataset')
plt.ylabel('Y Dataset')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_test
```

```python
## Lets implement simple linear regresion
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_test,regression.predict(X_test))
print(score)
```

```python
## Lets visulatize this model
plt.plot(X_train,regression.predict(X_train),color='r')
plt.scatter(X_train,Y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y Dataset")
```

```python
## Lets apply Ploynomial Transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy=PolynomialFeatures(degree=2,include_bias=True)
X_tarin_poly=ploy.fit_transform(X_train)
X_tset_poly=ploy.transform(X_test)
```

```python
X_tarin_poly
```

```python
# Inculde bias parameter
from sklearn.metrics import r2_score
regression_1=LinearRegression()
regression_1.fit(X_tarin_poly,Y_train)
Y_pred=regression_1.predict(X_tset_poly)
score_1=r2_score(Y_test,Y_pred)
print(score_1)
```

```python
print(regression_1.coef_)
```

```python
print(regression_1.intercept_)
```

```python
plt.scatter(X_train,regression_1.predict(X_tarin_poly))
plt.scatter(X_train,Y_train)
```

```python
ploy=PolynomialFeatures(degree=3,include_bias=True)
X_tarin_poly=ploy.fit_transform(X_train)
X_tset_poly=ploy.transform(X_test)
```

```python
X_tarin_poly
```

```python
# Inculde bias parameter
from sklearn.metrics import r2_score
regression_1=LinearRegression()
regression_1.fit(X_tarin_poly,Y_train)
Y_pred=regression_1.predict(X_tset_poly)
score_1=r2_score(Y_test,Y_pred)
print(score_1)
```

```python
## Prediiction of new dataset
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_ploy=ploy.transform(X_new)
X_new_ploy
```

```python
Y_new=regression_1.predict(X_new_ploy)
plt.plot(X_new,Y_new,"r-",linewidth=2,label="predictions")
plt.plot(X_train,Y_train,"b.",label="Trainig Points")
plt.plot(X_new,Y_new,"r-",linewidth=2,label="Testing Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

## Pipeline Concept

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new=np.linspace(-3,3,200).reshape(200,1)
    X_new_ploy=ploy.transform(X_new)
    poly_feature=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("poly_feature",poly_feature),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(X_train,Y_train)
    Y_pred_new=poly_regression.predict(X_new)
    ##poltting predicted line
    plt.plot(X_new,Y_pred_new,'r',label="Degree"+str(degree),linewidth=2)
    plt.plot(X_train,Y_train,"b.",linewidth=3)
    plt.plot(X_test,Y_test,"g.",linewidth=3)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend(loc="upper left")
    
    plt.axis([-4,4,0,10])
    plt.show()
    
```

```python
poly_regression(5)
```
